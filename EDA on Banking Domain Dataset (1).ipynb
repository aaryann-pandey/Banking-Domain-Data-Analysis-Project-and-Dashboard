```python
import pandas as pd
import numpy as np

less_10 = []
between_10_30 = []
greater_30 = []

for col in df.select_dtypes(include=[np.number]).columns:
    mean_val = df[col].mean()
    median_val = df[col].median()
    std_val = df[col].std()

    if std_val == 0:
        continue  # avoid division by zero

    diff = abs(mean_val - median_val)
    pct = diff / std_val * 100   # percentage difference

    if pct < 10:
        less_10.append(col)
    elif 10 <= pct <= 30:
        between_10_30.append(col)
    else:
        greater_30.append(col)

print("Columns with |mean - median| < 10% of std:")
print(less_10)

print("\nColumns with 10% ≤ |mean - median| ≤ 30% of std:")
print(between_10_30)

print("\nColumns with |mean - median| > 30% of std:")
print(greater_30)
```

# Insights of EDA:


## 1. Risk Weighting Distribution

Risk Weighting Level 2 contains the largest share of customers (over 41%), indicating that most users fall into a low-to-moderate risk bracket.

Platinum customers show a slightly higher proportion in Risk 1, implying they are more stable and less risky.

Silver customers show a slightly higher proportion in Risk 3 and Risk 5, indicating the presence of more moderate-to-high risk profiles in this group.

## 2. Loan Behavior

Over 52% of users have bank loan amounts in the lower range, suggesting that a majority of customers manage relatively smaller borrowing amounts.

Users with higher Risk Weighting levels tend to have higher loan amounts, reflecting a positive correlation between risk and liability.

Customers maintaining very high checking account balances are also likely to have high bank loan values, indicating active financial activity.

Customers who already take bank loans often take business lending loans as well, showing complementary borrowing behavior.

## 3. Age Characteristics

A significant portion of the dataset consists of older customers, with more than 72% above age 50.
This indicates a mature customer base with established financial behaviors.

## 4. Nationality

European nationality is the most represented among users.

## 5. Credit Card Behavior

Around 63% of customers have only one credit card, indicating conservative credit card usage.

## 6. Income & Its Impact

Higher estimated incomes (above 300,000) appear associated with higher risk levels, possibly due to larger transactions and loan activity.

Users belonging to higher income bands:

Maintain higher checking account balances than savings balances

Tend to have higher loan amounts

This suggests that higher-income individuals are more actively engaged in transactional banking and larger credit usage.

## 7. Deposits & Lending Relationships

Users with higher bank deposits tend to take larger business lending loans, showing that wealthier customers also engage in corporate or commercial borrowings.

Although customers with high deposits show strong financial health, their foreign currency deposits do not increase proportionally with local deposits—indicating foreign currency accounts are used selectively.

## 8. Checking–Savings Relationship

Customers who keep high checking account balances also tend to maintain high savings account balances.
This suggests an overall wealth-related behavior where affluent customers maintain liquidity across both accounts.

## 9. Properties Owned

The number of properties owned shows no meaningful correlation with banking behavior (deposits, loans, risk, or income).
Therefore, it does not appear to be a strong predictive or explanatory variable in this dataset.

## 10. Loyalty Classification

Loyalty tiers (Jade, Gold, Silver, Platinum) show very similar risk distributions, meaning:

Loyalty classification reflects customer engagement/value, not risk.

Risk is more influenced by income, loans, and account balances than loyalty status.
